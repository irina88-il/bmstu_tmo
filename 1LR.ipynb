Этот датасет содержит медицинские данные, связанные с сердечными заболеваниями. Он включает различные признаки, которые могут использоваться для прогнозирования наличия или отсутствия болезни сердца.

Описание признаков:
Age: возраст пациента
Sex: пол пациента (1 = мужчина, 0 = женщина)
Chest Pain Type: тип боли в груди (4 значения):
0: тип 0
1: тип 1
2: тип 2
3: тип 3
Resting Blood Pressure: артериальное давление в состоянии покоя (мм рт. ст.)
Serum Cholestoral: уровень холестерина в сыворотке (мг/дл)
Fasting Blood Sugar: уровень сахара натощак > 120 мг/дл (1 = true, 0 = false)
Resting Electrocardiographic Results: результаты ЭКГ в состоянии покоя (0, 1, 2)
Maximum Heart Rate Achieved: максимальная достигнутая частота сердечных сокращений
Exercise Induced Angina: стенокардия, вызванная физической нагрузкой (1 = да, 0 = нет)
Oldpeak: депрессия ST сегмента, индуцированная физической нагрузкой относительно состояния покоя
The Slope of the Peak Exercise ST Segment: наклон пика сегмента ST при физической нагрузке
Number of Major Vessels: количество основных сосудов (0-3), окрашенных флюороскопией
Thal: талассемия (0 = нормальный, 1 = фиксированный дефект, 2 = обратимый дефект)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('heart.csv')
```

```python
data.head()
```

```python
data.head()
```

```python
missing_values = data.isnull().sum()
print("Пропуски в данных:")
print(missing_values)
```

```python
# Визуальное исследование датасета
plt.figure(figsize=(10, 6))
sns.histplot(data['age'], bins=30, kde=True)
plt.title('Распределение возраста')
plt.xlabel('Возраст')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='sex', data=data)
plt.title('Распределение пола')
plt.xlabel('Пол (1 = мужчина, 0 = женщина)')
plt.ylabel('Частота')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='cp', data=data)
plt.title('Распределение типа боли в груди')
plt.xlabel('Тип боли в груди')
plt.ylabel('Частота')
plt.show()
```

```python
# Корреляционная матрица
correlation_matrix = data.corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица признаков')
plt.show()
```
